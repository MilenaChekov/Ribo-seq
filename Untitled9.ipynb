```python
counts = df.drop(columns=['cor_norm', 'cor_tumor', 'geneID', 'geneSymbol'])
```

```python
# DESeq2
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter
```

```python
from rpy2.robjects.packages import importr

DESeq2 = importr("DESeq2")
```

```python
counts_RNA = counts[counts.columns[counts.columns.str.contains('RNA')]]

# Design matrix
design_RNA = pd.DataFrame({
    "condition": [sample_name.split("-")[1] for sample_name in counts_RNA.columns]
    }, index=counts_RNA.columns)

design_RNA.head()
```

```python
with localconverter(ro.pandas2ri.converter):
  countData = ro.conversion.py2rpy(counts_RNA)
  colData = ro.conversion.py2rpy(design_RNA)

dds = DESeq2.DESeqDataSetFromMatrix(countData=countData, colData=colData, design=ro.Formula("~ condition"))
dds = DESeq2.DESeq(dds)
```

```python
# Getting the results
to_dataframe = ro.r("function(x) data.frame(x)")
dds_results = to_dataframe(DESeq2.results(dds))

with localconverter(pandas2ri.converter):
    dds_results = ro.conversion.rpy2py(dds_results)
    dds_results.index = genes
```

```python
dds_results.sort_values(by='padj').head()
```

```python
p_tr = 0.05

fig, ax = plt.subplots(figsize=(5, 5), dpi=150)
xlim = max(-min(dds_results.log2FoldChange), (max(dds_results.log2FoldChange))) + 0.5

sns.scatterplot(
    x=dds_results.log2FoldChange,
    y=-np.log10(dds_results.padj),
    ax=ax,
    s=3,
    linewidth=0,
)

ax.plot(
    [-xlim, xlim],
    [-np.log10(p_tr), -np.log10(p_tr)],
    color="red",
)

sns.scatterplot(
    x=dds_results[dds_results.padj < p_tr].log2FoldChange,
    y=-np.log10(dds_results[dds_results.padj < p_tr].padj),
    ax=ax,
    s=5,
    linewidth=0,
    color="red",
)

ax.grid(alpha=0.3)

ax.set_xlim(-xlim, xlim)
ax.set_xlabel("log$_2$FoldChange")
ax.set_ylabel("log$_{10}$FDR")
```

```python
# Design matrix
design = pd.DataFrame({
    "condition": [sample_name.split("-")[1] for sample_name in counts.columns],
    "experiment": [sample_name.split("-")[2] for sample_name in counts.columns],
}, index=counts.columns)

design.head()
```

```python
with localconverter(ro.pandas2ri.converter):
  countData = ro.conversion.py2rpy(counts)
  colData = ro.conversion.py2rpy(design)

dds = DESeq2.DESeqDataSetFromMatrix(countData=countData, colData=colData, design=ro.Formula("~ condition + experiment + experiment:condition"))
dds = DESeq2.DESeq(dds)
```

```python
# Getting the results
to_dataframe = ro.r("function(x) data.frame(x)")
dds_results = to_dataframe(DESeq2.results(dds))

with localconverter(pandas2ri.converter):
    dds_results = ro.conversion.rpy2py(dds_results)
    dds_results.index = genes
```

```python
ann = pd.read_csv('GCF_000005845.2_ASM584v2_genomic.gff', comment='#', sep='\t', names=['seqname','source','feature','start','end','score','strand','frame','attribute'])
ann = ann.loc[ann['feature']=='gene']
ann
```

```python
wig_f_d = wig_f['counts'].to_dict()
wig_r_d = wig_r['counts'].to_dict()
```

```python
ann_f = ann.loc[ann['strand'] == '+']
ann_f['length'] = ann_f['end'] - ann_f['start']
ann_f
```

```python
from datetime import datetime
start_time = datetime.now()


gene = ann_f.iloc[0]
gene = dict.fromkeys(range(gene['start'], gene['end']+1))
for i in gene:
  gene[i] = wig_f_d.get(i,0)

fig, ax = plt.subplots(figsize=(15,5))
sns.scatterplot(ax=ax, x=gene.keys(), y=gene.values())

end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```

```python
import scipy.stats as st
print(st.mode(ann_f['length']))
print(st.mode(ann_f.loc[ann_f['length'] > 100]['length']))
```

```python
genes = ann_f.loc[ann_f['length'] == 75]
dist = dict.fromkeys(range(75),0)
for i in genes['start']:
  for j in range(75):
    dist[j] += wig_f_d.get(i+j, 0)/28
fig, ax = plt.subplots(figsize=(15,5))
sns.scatterplot(ax=ax, x=dist.keys(), y=dist.values())
```

RPKM - Reads Per Kilobase Million

TPM - Transcripts Per Kilobase Million

TPM позволяет сравнивать разные образцы

```python
ann_f['counts'] = 0
for i in ann_f.index:
  for j in range(ann_f.loc[i, 'start'], ann_f.loc[i, 'end']+1):
    ann_f.loc[i, 'counts'] += wig_f_d.get(j, 0)
ann_f
```

```python
ann_f['RPKM'] = (ann_f['counts']*1000000000)/(np.sum(ann_f['counts'])*ann_f['length'])
ann_f['TPM'] = ann_f['RPKM']/np.sum(ann_f['RPKM'])*1000000
ann_f
```
